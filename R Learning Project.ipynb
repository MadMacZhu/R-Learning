```r
# Data Modeling: Hierarchical Clustering
library(datasets)
pacman::p_load(pacman, dplyr)
head(mtcars)
cars <- mtcars[ , c(1:4, 6:7, 9:11)] # First blank means all rows,
                                     # second argument select columns
head(cars)

# Compute and plot Clusters
hc <- cars %>%    # Get cars data
      dist %>%    # Compute distance/dissimilarity matrix
      hclust      # Compute hierarchical clusters

plot(hc)          # Plot dendrogram

# Add boxes to plot
rect.hclust(hc, k = 2, border = "gray")
rect.hclust(hc, k = 3, border = "blue")
rect.hclust(hc, k = 4, border = "green4")
rect.hclust(hc, k = 5, border = "darkred")
```

```r
# Modeling Data: Principal Components (aka. Dimensionality Reduction)
library(datasets)
pacman::p_load(pacman, dplyr)
cars <- mtcars[ , c(1:4, 6:7, 9:11)]

# Compute PCA
pc <- prcomp(cars,
             center = TRUE,    # Centers means to 0 (optional)
             scale = TRUE)     # Sets unit variance (helpful)
# OR to specify variables
pc <- prcomp(~ mpg + cyl + disp + hp + wt + qsec + am + gear + carb,
             data =  mtcars,
             center = TRUE,
             scale = TRUE)
summary(pc)

# Screeplot for number of components
plot(pc)

# Get standard deviations and rotation
pc

# See how cases load on PCs
predict(pc) %>% round(2)

# Biplot of the first two components
biplot(pc)
```

```r
# Modeling Data: Regression
library(datasets)
head(USJudgeRatings)
data <- USJudgeRatings

# Define variable groups
x <- as.matrix(data[-12])
y <- data[ , 12]

# Regression using variable groups
reg1 <- lm(y ~ x)

# Or specify variables individually
reg2 <- lm(RTEN ~ CONT + INTG + DMNR + DILG + CFMG + DECI + PREP + 
           FAMI + ORAL + WRIT + PHYS, 
           data = USJudgeRatings)

# Results
reg1
summary(reg1)
reg2
summary(reg2)
```

```r
# More summaries
anova(reg1)              # Coefficients with inferential tests
coef(reg1)               # Coefficients (same as reg1)
confint(reg1)            # CI for coefficients
resid(reg1)              # Residuals case-by-case
hist(residuals(reg1))    # Histogram of residuals
```

```r
# Additional Models
pacman::p_load(lars, caret)

# Conventional stepwise regression
stepwise <- lars(x, y, type = "stepwise")

# Stagewise: like stepwise but has better generalizability
forward <- lars(x, y, type = "forward.stagewise")

# LAR: Least angle regression
lar <- lars(x, y, type = "lar")

#LASSO: Least absolute shrinkage and selection operator
lasso <- lars(x, y, type = "lasso")

# Compare R2 for new models
r2comp <- c(stepwise$R2[6], forward$R2[6], lar$R2[6], lasso$R2[6]) %>%
          round(2)
names(r2comp) <- c("stepwise", "forward", "lar", "lasso")
r2comp
```
